```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("flower.jpg")
imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
h, w= imgRGB.shape[:2]
```

```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh gốc (nhớ chuyển BGR sang RGB để hiển thị đúng màu với matplotlib)
img = cv2.imread("flower.jpg")
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# --- Phóng to ảnh (ví dụ: phóng to 1.5 lần theo cả chiều ngang và dọc) ---
# Sử dụng fx và fy để chỉ định hệ số tỉ lệ [cite: 41]
img_scaled_up = cv2.resize(imgRGB, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_LINEAR)

# --- Thu nhỏ ảnh (ví dụ: thu nhỏ còn một nửa) ---
# Cách 1: Sử dụng fx và fy
img_scaled_down_fx_fy = cv2.resize(imgRGB, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_AREA)

# Cách 2: Chỉ định kích thước đầu ra
# new_width = imgRGB.shape[1] // 2
# new_height = imgRGB.shape[0] // 2
# img_scaled_down_size = cv2.resize(imgRGB, (new_width, new_height), interpolation=cv2.INTER_AREA)

# Hiển thị ảnh gốc và ảnh đã thay đổi kích thước
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(imgRGB)
plt.title('Ảnh Gốc')

plt.subplot(1, 3, 2)
plt.imshow(img_scaled_up)
plt.title('Ảnh Phóng To (x1.5)')

plt.subplot(1, 3, 3)
plt.imshow(img_scaled_down_fx_fy)
plt.title('Ảnh Thu Nhỏ (x0.5)')

plt.show()
```
